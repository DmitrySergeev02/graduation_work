```python
import spacy
from fastcoref import spacy_component

nlp = spacy.load("en_core_web_sm")
nlp.add_pipe("fastcoref",config={'model_architecture': 'LingMessCoref', 'model_path': 'biu-nlp/lingmess-coref', 'device': 'cuda'})

def clean_text(text):
    text = remove_special_characters(text)
    doc = nlp(text,component_cfg={"fastcoref": {'resolve_text': True}})
    text = doc._.resolved_text
    return text
```

```python
import re

def remove_special_characters(text):
    regex = re.compile(r'[\n\r\t]')
    clean_text = regex.sub(" ", text)
    return clean_text
```

```python
from pynytimes import NYTAPI
from newspaper import Article
import json

topics = ['Politics','Arts','Books','Business','Culture','Education','Financial','Media','Movies','Museums','Science','Sports','Technology','Television','Society']

data=[]

for topic in topics:
  nyt = NYTAPI("GYpoqaE7bHCgbQf3Zytxo7j80YnsQKvJ", parse_dates=True)
  articles = nyt.article_search(
      results = 50,
      options = {
          "sort": "newest",
          "news_desk": [
              topic
          ],
      }
  )
  for story in articles:
      if story['document_type']!='article':
          continue
      article = Article(story['web_url'])
      article.download()
      article.parse()
      article_text = article.text
      article_text = clean_text(article_text)
      data.append({"url": story['web_url'],"text": article_text})
      print(story['web_url'])
      print('-------------------------------')

with open('data.json', 'w') as f:
    json.dump(data, f)
```
